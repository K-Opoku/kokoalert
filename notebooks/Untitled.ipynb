```python
from pathlib import Path
from src.preprocess import file_to_spectrograms, check_recording_quality
from src.anomaly_detector import load_autoencoder, is_anomalous
import numpy as np

model, threshold = load_autoencoder()
print(f"Model loaded. Threshold: {threshold}\n")

ghana_folder = Path("../data/raw/ghana/")
clips = [f for f in ghana_folder.iterdir()
         if f.suffix.lower() in (".wav", ".ogg", ".mp3", ".m4a", ".flac")]

print(f"Found {len(clips)} clips\n{'─'*50}")

for clip in clips:
    quality = check_recording_quality(str(clip))
    if not quality["usable"]:
        print(f"❌ {clip.name} — unusable ({quality['reason']})")
        continue

    specs = file_to_spectrograms(str(clip))
    results = [is_anomalous(model, s, threshold) for s in specs]

    probs = [r["probability"] for r in results]
    avg_prob = np.mean(probs)
    anomaly_votes = sum(1 for r in results if r["is_anomalous"])
    verdict = "🔴 SICK" if anomaly_votes > len(results) / 2 else "✅ HEALTHY"

    print(f"{clip.name}")
    print(f"  Duration quality: {quality['duration']:.1f}s")
    print(f"  Windows: {len(specs)}")
    print(f"  Avg P(sick): {avg_prob:.3f}")
    print(f"  Anomaly votes: {anomaly_votes}/{len(results)}")
    print(f"  Verdict: {verdict}")
    print()
```
