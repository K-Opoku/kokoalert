```python
import librosa
from pathlib import Path

```

```python
for split in ["healthy", "sick", "noise"]:
    durations = []
    for f in Path(f"data/raw/bowen/{split}").glob("*.wav"):
        d = librosa.get_duration(path=f)
        durations.append(d)
    
    if durations:
        windows = sum(int((d-5)/2+1) for d in durations if d >= 5)
        print(f"\n{split.upper()}: {len(durations)} clips")
        print(f"  Mean duration: {sum(durations)/len(durations):.1f}s")
        print(f"  Total audio: {sum(durations)/60:.1f} mins")
        print(f"  Estimated training windows: {windows}")
```
